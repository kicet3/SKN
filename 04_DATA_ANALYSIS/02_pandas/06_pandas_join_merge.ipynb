```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame({
    'n_id':[1,2,3,4],
    'name':['토끼','판다','코알라','양']
})
df2 = pd.DataFrame({
    's_id':[1,2,3,5],
    'score':[100,95,88,71]
})
```

```python
pd.merge(df1,df2,left_on='n_id',right_on='s_id')
```

```python
df5 = pd.DataFrame({
    'name':['구준표','윤지후','소이정','프린스송'],
    'age':[18,18,18,18]
})
df6 = pd.DataFrame({
    'name':['금잔디'],
    'age':[17]
})

pd.concat([df5,df6],axis=0).reset_index(drop=True)
```

```python
pd.concat([df5,df6],axis=1).reset_index(drop=True)
```

```python
df = pd.DataFrame({
    'name': ['홍길동', '신사임당', '이순신'],
    'math': [80, 99, 100],
    'english': [34, 58, 48],
    'science': [99, 55, 77],
})
```

```python
pd.melt(df,
        id_vars=['name'],
        value_vars=['math','english','science'],
        var_name='subject',
        value_name='score'
        )
```
