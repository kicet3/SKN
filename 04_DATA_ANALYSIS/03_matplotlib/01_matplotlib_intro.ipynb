##### Figure | Axes
- Figure : 전체 그림이 그려지는 컨테이너
- Axes : 그래프가 실제 그려지는 영역 (좌표,데이터)

```python
fig = plt.figure(figsize=(6,4))

ax = plt.gca()
ax.plot([1,2,3],[4,5,6])
print(ax)

plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=3,figsize=(10,6))
print(fig)
print(ax)
plt.show()
```

```python
x = np.array(range(11))
y = x*2
plt.plot(x,y)
plt.show()
```

```python
x = pd.Series(x)
y = pd.Series(y)
plt.plot(x,y)
plt.show()
```

```python
# 4. DataFrame
x = pd.DataFrame({'x':x})
y = pd.DataFrame({'y':y})

plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(3,2))
x = np.arange(1,101)
y = np.linspace(3,200,100)
y2 = np.linspace(3,300,100)

plt.plot(x,y,label='시간')
plt.plot(x,y2,label='공간')

# loc 범례 위치 (기본 값 :best)
# - (upper | lowe) (left | right)
# - center

plt.legend(loc='lower right')
plt.show()
```

##### 그래프 여러개 그리기


```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
x1 = np.arange(1,11)
y1 = x1*2
x2 = np.arange(11,20) 
y2 = x2*2

ax[0].plot(x1,y1,label='선',color='red')
ax[0].set_xlabel('x1')
ax[0].set_ylabel('y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()

ax[1].bar(x2,y2,label='바',color='yellowgreen')
ax[1].set_xlabel('x2')
ax[1].set_ylabel('y2')
ax[1].set_title('연두색')
ax[1].legend()
ax[1].grid()
```
