```python
def drop_column(df,column_list):
    for column_name in column_list:
        df = df.drop(column_name,axis=1)
    return df
```

```python
columns = [
    "host name",
    "country",
    "country code",
    "calculated host listings count",
    "license",
    "house_rules",
    "last review",
    "instant_bookable",
    "long",
    "lat",
    "neighbourhood group",
    "NAME",
    "host id",
    "host_identity_verified"
]

airbnb_df = drop_column(airbnb_df,columns)
airbnb_df.info()
```

```python
airbnb_df['neighbourhood'].unique()
```

```python
airbnb_df['price'] = airbnb_df['price'].replace(r'[\$,]', '', regex=True).astype(float)
airbnb_df['service fee'] = airbnb_df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
airbnb_df.describe()
```

```python
def fillna(df,columns,default=0):
    for column in columns:
        df[column] = df[column].fillna(default)
    return df

def dropna(df,column_list):
    df = df.dropna(subset=column_list,axis=0)
    return df

def cleanedData(df,columns,value,type='over'):
    for column in columns:
        if type == 'over':
            df = df[df[column] < value ]
        elif type == 'under':
            df = df[df[column] > value ]
    return df
```

```python
airbnb_df = fillna(airbnb_df,['price'],airbnb_df['price'].mean())
airbnb_df = cleanedData(airbnb_df,['price'],50,'under')
airbnb_df = cleanedData(airbnb_df,['price'],1000,'over')
airbnb_df = cleanedData(airbnb_df,['availability 365'],value=0,type='under')
airbnb_df = cleanedData(airbnb_df,['availability 365'],value=366,type='over')
airbnb_df = fillna(airbnb_df,['minimum nights'],1)
airbnb_df = dropna(airbnb_df,['availability 365'])
airbnb_df['price'] = airbnb_df['price'].astype(int) * 1400
airbnb_df = fillna(airbnb_df,['service fee'])
airbnb_df['service fee'] = airbnb_df['service fee'].astype(int) *1400
```
